**Object Oriented Programming (OOP)-**

A class is a blueprint/template for creating objects.

ex - Class = Design of a House

Object = Actual House

-> __init__() is a special method called constructor.

It automatically runs whenever an object is created.

**self -** Refers to the current object being created.

**name -** Receives the value passed during object creation.

If we create an method and access to it then we can only access to it then whereas when we create a class then it contains many methods in it so we don't need to access each method again and again and we can access all the methods at once as per our requirement by accessing the class at once.

A constructor (__init__) is not mandatory in Python. If no constructor is defined, Python automatically provides a default constructor that creates the object.

No constructor?         → Fine

No methods?             → No self needed

Instance method exists? → self needed

```python
# example
class himani:
  def __init__(self,name):  # self , name - parameters
    self.name = name  # stores data -> self is like an empty container
    # instance variable

    #self is used here so can we can use name easily

    #internally -> self = r
    #name = hello

  def show(self): #creates a method
    print(self.name)

r = himani("hello")  # creates an object r # hello -> argument
r.show()
```

```python
#example 4
class student:
  def __init__(self,*args):
    self.data = args

    # *args -> positional arguments -> tuple
    #collects all keyword arguments into a tuple

  def users(self):
    for i in self.data[0]:
      print(i)

  def details(self):
    for i in self.data[1]:
      print(i, "=", self.data[1][i])

s = student(["dheeraj","kunal","harsh","praveen"],{"address":"kukas","college":"arya","loc":"jaipur"})
s.users()
s.details()
```

```python
class student:
  def __init__(self,name,age):
    print(type(name))
    self.name = name

  def data(self):
    print(self.data["name"])

d = {"name":"himani","age":20}
s = student(**d)

# **d -> It unpacks the dictionary into keyword arguments
# internally -> s = student(name="hello", age=20)
```

```python
def stu(*a):
  print(a)

stu([10,20,40])

#The comma is there because Python shows a tuple with one element as:
#(element,)

#*a in function definition
#→ Collect arguments into a tuple

#*a in function call
#→ Unpack list/tuple into separate arguments
```
