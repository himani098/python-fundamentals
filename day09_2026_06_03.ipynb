**slicing** - Used to take a part of a string/list using indexes.
ex - s[:]

**split**- Used to break a string into multiple pieces and return a list. ex - s.split()

in numpy we can split an array into multiple sub-arrays

**slicing on a NumPy array:-**

[start : stop : step]

start → Where to begin -> beginning index

stop  → Where to stop (not included) -> ending index

step  → How much to jump -> bydefault 1 if nothing

After taking one element, how many indexes should Python move forward to get the next element?

**2D NumPy slicing.**

array[rows, columns]

before comma -> row , after comma -> column

```python
# example split

import numpy
n = numpy.arange(12).reshape(4,3)
print(n)
print("\n")
print(n[0:1:1])
print("\n")
print(n[: ,:2])
#Rows    → :   → all rows
# Columns → : 2 → start at column 0 & stop before column 2
```

```python
import numpy
n = numpy.arange(12).reshape(4,3)
print(n)
print("\n")
print(n[0:4:,2])
print("\n")
print(n[0:4,2:4])
print("\n")
print(n[:,-1])  # -1 --> last column

#Single index  (-1)  → 1D output
# Slice         (-1:) → 2D output
```

```python
# example -> print 4,7

import numpy
n = numpy.arange(12).reshape(4,3)
print(n)
print("\n")
print(n[1:3,1:2])
print("\n")
# or
print(n[1:3,1])  #-> column with 1 index
```
