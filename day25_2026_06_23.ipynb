```python
#data preprocessing
import pandas as pd
data = {
    'Name':['Ram','Kamal','Ajay'],
    'Age':[25,None,32],
    'Salary':[5000,6000,None]
}
df = pd.DataFrame(data)
print("original DataFrame")
print(df)
print("\n")

# null find out | fillna | dropna
print(df.isnull().sum())
print("\n")
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
print("\n")
print(df.dropna())
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
import pandas as pd

data = {
    'Name':['Ram','Kamal','Ajay','Neetu','Kavi','Sapu'],
    'Age':[25,None,32,20,21,22],
    'Salary':[5000,6000,None,7000,8000,9000],
    'Gender':['male','male','male',None,'female','female']
}
df = pd.DataFrame(data)
print("original dataframe")
print(df)
print("\n")

#label encoder
le = LabelEncoder()
df_label = df.copy()
df_label['Gender'] = df['Gender'].fillna('unkown')
df_label['Gender-encoder'] = le.fit_transform(df_label['Gender'])
#assigns a specific number to same values

print("After label encoding:")
print(df_label)
```

```python
#one hot encoding
# no need of sklearn in this
import pandas as pd
data = {
    'colors':['red','blue','green','red','blue']
}
df = pd.DataFrame(data)
print("original data")
print(df)
print("\n")

#one hot encoding
encoded_df = pd.get_dummies(df,columns = ['colors'])

print("after one hot encoding")
print(encoded_df)
```
