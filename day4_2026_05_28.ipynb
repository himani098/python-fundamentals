**Lambda Function-**
A lambda function is a small anonymous (nameless) function written in a single line.

```python

#Lambda Functions [lambda arguments : expression]
show = lambda x: x
print(show(2000))


sum = lambda x,y: x+y
print(sum(10,20))

#expression is what you want to print or perform any operation
```

```python
a = lambda x:x[0]
print(a([10,20,30,40]))


#question
a = lambda *x: [i for i in x]
print(a(10,20,30))
```

```python
#List Comprehensions [expression for item in iterable]
#Expression
#What you want to store in the new list.
#Item
#The variable that takes values one by one.
#Iterable
#Anything you can loop over.
#list comprehension is a short way to create a list using a loop.

print([i for i in range(5)])
```

```python
#example
l = [10,20,30,40,50,60]
print([l[i] for i in range(len(l))])
```

```python
#question
l = [10,20,[30,40,50,60]]
print([l[2][i] for i in range(len(l[2]))])
print([l[-1][i] for i in range(len(l[-1]))])
```

```python
#list
l = [10,20,30,40,50,"hello",True]
print(l)
print(l[0])
print(l[-1])
# [-1] for last index value
print(l[-2])
# [-1] for second last index value
```

**Dictionaries**

Key-Value Operations

Nested Dictionaries

```python
# dict
d = {"name":"himani","age":20}
print(d['name'])
print(d.keys())
print(d.values())
print("\n")

for i in d.keys():
  print("key=",i)
  print("value=",d[i])
```
