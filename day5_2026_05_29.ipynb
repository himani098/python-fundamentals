```python
d = {"name":"himani"}
d.update({"name":"tia"})
d['name'] = "tanya"
print(d)

# dictionary works on keys and not indexing
```

```python
student = {
    "name": "Himani",
    "age": 20,
    "college": "Arya"
}
#get() -> Gets a value using a key.
print(student.get("name"))

#keys() -> Returns all keys.
print(student.keys())

#values() -> Returns all values.
print(student.values())

#items() -> Returns all key-value pairs.
print(student.items())

#update() -> Adds or updates values.
student.update({"city": "Jaipur"})
print(student)

#copy() -> Creates a copy of the dictionary.
student2 = student.copy()
print(student2)

#clear() -> Removes everything.
student.clear()
print(student)

#del -> Deletes a key-value pair.
#Deletes a key or the whole dictionary.
del student2["age"]
print(student2)

del student2
#print(student2)
# this line now gives error bcz this dictionary no longer exists now
```

JSON is a data format:-

Rules:
1) Uses true, false, null
2) Keys must be in double quotes " "
3) Used for data exchange between server and application, for APIs , files

Python dictionary is a Python data type:-

Rules:

1) Uses True, False, None
2) Keys can be strings, numbers, tuples, etc.
3) Can be used directly in Python code, programs



**Nested List**

Slicing

```python
l = [10,20,30,["hello","hello1","hello2",[True,False]]]

# general rule for slicing:-
# l[start:end]
# Start index is included ✅
# End index is excluded ❌

l1 = l[:3]
print(l1)
# Start from index 0
# Go up to index 3 (not including 3)

print(l[2:])
# Start from index 2
# Go till the end

print(l[3][1:3])
# for printing values btw two nested lists

print(l[-3:-1])
print(l[-3:-3])
print(l[-3:-2])
print(l[-1:-3])

# -3,-2,-1  # <-
```

**Map**(Transform values), **Filter**(Select values), **Reduce** (Combine values):-

These are Python functions used to process data without writing full loops.
They do not modify the original list.
They create a new result

map    → Modify every element

filter → Select some elements

reduce → Produce one final value

```python
# map ( atleast 2 arguments necessary)
# map() -> Used to apply a function to every element.
# map(function, iterable)
# An iterable is anything that can be looped over using for.
# ex - tuple,list,dictionary,string,range

l = [1,2,3,4]
result = map(lambda x: x*2, l)
print(list(result))
# list(result) -> Convert the map object into a list.
```

```python
# function
def square(x):
    return x * x

l = [10,20,30]
l1 = list(map(square,l))
print(l1)
```

```python
l = [10,20,30]
l1 = list(map(lambda x:x*x,l))
print(l1)

# alternative | without in built function
l1 = [] # length = 1
for i in range(len(l)):
  l1.append(l[i] * l[i])
print(l1)
```

```python
#filter(atleast 2 arguments necessary)
#filter() -> Used to keep only elements that satisfy a condition.
#filter(function, iterable)
#Keeps elements for which the function returns True.

l = [1,2,3,4,5,6]
result = filter(lambda x: x%2==0, l)
print(list(result))
```

```python
def helo(x):
  return x.endswith('a')
  # return x.startswith('a')

l = ["apple","banana","cat","dog"]
l1 = list(filter(helo,l))
print(l1)

#example | alternative of filter
l1 = []
for i in l:
  if 'a' == i[-1]:
    l1.append(i)
print(l1)
```

```python
#HW
# to convert all the letters to uppercase

l = ["apple","banana","cat","dog"]
l1 = list(map(lambda x : x.upper() ,l))
print(l1)
```

**from functools import reduce:-**

Import the reduce() function from the functools library so that I can use it in my program.

from      → take from

functools → library name

import    → bring into
program

reduce    → function name


Some functions are built into Python:print() , input() , len() -> You can use them directly.

But reduce() is stored inside a library called functools, so Python needs to know where to find it.


**Import only reduce**

from functools import reduce

reduce(...)

**Import entire library**

import functools

functools.reduce(...)

```python
#reduce(atleast 2 arguments necessary)
#reduce() -> used to combine all elements into a single value.

from functools import reduce

l = [1,2,3,4]
result = reduce(lambda x,y: x+y, l)
print(result)
# 1 + 2 + 3 + 4 = 10
```


**Exception Handling**

try, except, finally

try:
    # risky code

except:
    # runs if error occurs

else:
    # if no error

finally:
    # always runs

```python
# try except
try:
  num1  = 10
  num2 = 5
  print(num1/num2)
except:
  print("hello except")
```

```python
# try except
try:
  num1 = 10
  num2 = 0
  print(num1/num2)
except:
  print("hello except")

#without exception handling
#ZeroDivisionError: division by zero
#Program stops immediately.
```

```python
try:
    num = int(input("Enter number: "))

except ValueError:
    print("Invalid input")

# ValueError is not a name that we created.
# It is a built-in exception (error type) already provided by Python.
```

```python
try:
    a = 10
    b = 2
    print(a / b)

except ZeroDivisionError:
    print("Error")

else:
    print("Division successful")

#else
# Runs only if no error occurs.
```

```python
# try except and final
try:
  num1 = 10
  num2 = 5
  print(num1/num2)
except:
  print("hello except")
finally:
  print("hello finally divide")

#finally
#Runs whether an error occurs or not.
```
