**Operation on Rows**

1) print(df.loc[row,"col name"]) -> label based (get single value)
2) print(df.loc[row]) -> label based (get complete row)
3) print(df.loc[[0, 2]]) -> get multiple rows
4) print(df.loc[[0, 2], ["col1", "col2"]]) multi rows and cols
5) print(df.iloc[row_position, column_position])
6) print(df.iloc[row_position])
7) print(df.iloc[[row1, row2]])
8) print(df.iloc[[row1, row2], [col1, col2]])
9) print(df.iloc[[0, 2], [0]])
10) print(df.iloc[1, [0, 2]])
11) print(df.iloc[0:3])

loc → Location by LABEL (name)

uses :- Row labels

Column labels

iloc → Integer Location (position/index number)

uses :- Row positions

Column positions

```python
import pandas as pd
d = {
    "name":["vishal","virat","vineet"], # 3 rows
    "salary":[100,200,300] # 3 rows
}
df = pd.DataFrame(data=d)
print(df)
print("\n")
print(df.loc[2,"name"])
print("\n")
print(df.iloc[2,0])
print("\n")
print(df.loc[1])
print("\n")
print(df.loc[[0, 2]])
print("\n")
print(df.loc[[0, 2], ["name", "salary"]])
print("\n")
print(df.iloc[1, 0])
print("\n")
print(df.iloc[1])
print("\n")
print(df.iloc[[1, 2]])
print("\n")
print(df.iloc[[1, 2], [0, 1]])
print("\n")
print(df.iloc[[0, 2], [0]])
print("\n")
print(df.iloc[1, [0, 1]])
print("\n")
print(df.iloc[0:3])
```

```python
import pandas as pd
d = {
    "name":["vishal","virat","vineet"], # 3 rows
    "salary":[100,200,300] # 3 rows
}
df = pd.DataFrame(data=d)
print(df)
print("\n")
print(df.loc[0:3])
print("\n")

#sub data get
df1 = df.iloc[0:2,[0]] #uses python slicing --> end value excluded
print(df1)
print("\n")

df2 = df.loc[0:2,["name"]]  #uses label slicing --> end value included
print(df2)
```

**Filter DataFrame**

1) Filter DataFrame -> Ex. df[df["col name"] > 22]
2) Filter Using Greater Than (>)
3) df[df["column"] > value]
4) Filter Using Less Than (<)
5) df[df["column"] < 22]
6) Filter Using Equal To (==)
7) df[df["column"] == 22]
8) Filter Using Not Equal To (!=)
9) df[df["Age"] != 22]
10) Filter String Columns
11) df[df["column"] == "Sara"]
12) Multiple Conditions (AND)
13) df[(df["col1"] > 22) & (df["col2"] > 55000)]
14) Multiple Conditions (OR)
15) df[(df["col1"] > 23) | (df["col2"] > 55000)]
16) Using query()
17) df.query("col1 > 22")

```python
import pandas as pd
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/student_data.json"
df = pd.read_json(url)
print(df)
print("\n")
print(df[df["english"] >= 90])
print("\n")
print(df[df["name"] == "Sara"])
print("\n")
print(df[(df["maths"] > 90) & (df["english"] > 90) & (df["gender"] == "Male")])
print("\n")
print(df[(df["maths"] > 90) & (df["english"] > 90) & (df["gender"] == "Male")]["name"])
print("\n")
print(df.query("maths > 60"))
```

```python
import pandas as pd
data = {
    "Emp_ID": [101,102,103,104,105,106],
    "Name": ["Amit","Riya","Raj","Sara","John","Neha"],
    "Department": ["IT","HR","Finance","IT","Sales","HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)
print("\n")
print(df[df["Experience"] > 3])
print("\n")
print(df[df["Experience"] < 3])
print("\n")
print(df[df["Experience"] == 3])
print("\n")
print(df[df["Experience"] != 3])
print("\n")
print(df[df["Name"] == "Raj"])
print("\n")
print(df[(df["Salary"] > 40000) & (df["Experience"] > 3)])
print("\n")
print(df[(df["Salary"] > 40000) | (df["Experience"] > 3)])
print("\n")
print(df.query("Experience > 3"))
```
