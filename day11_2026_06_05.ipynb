**Array Reshaping**

1) reshape()
2) flatten() -> create copy | shallow copy
3) ravel() -> return original | deep copy
4) transpose -> .T -> reverse

```python
#reshape array
import numpy as np
arr = np.arange(12)
up_arr = np.reshape(arr,(3,4))
print(up_arr)
```

```python
# example 2d
import numpy as np
arr1 = np.arange(24)
up_arr1 = np.reshape(arr1,(12,2))
print(up_arr1)
```

```python
# example 3d
import numpy as np
arr2 = np.arange(24)
up_arr2 = np.reshape(arr2,(3,2,4))
print(up_arr2)
```

**flatten()** → creates a copy

Returns a copy (new independent data).

makes flat

Converts a multi-dimensional array into a 1D array.

**ravel()**   → creates a view (more memory efficient)

Returns a view (shares original data).

Also converts to a 1D array.
It does not create a new array.

It points to the same data as the original array.

```python
#flatten -> create copy then works

# example 2d
import numpy as np
arr = np.arange(12).reshape(3,4)
print(arr)
print("\n")
up_arr = arr.flatten()
up_arr[0] = 100
print(up_arr)
print("\n")
print(arr)
```

```python
#example 3d
import numpy as np
arr1 = np.arange(24).reshape(3,2,4)
print(arr1)
print("\n")
up_arr1 = arr1.flatten()
up_arr1[0] = 100
print(up_arr1)
print("\n")
print(arr1)
```

```python
#ravel -> works on original array

# example 2d
import numpy as np
arr = np.arange(14).reshape(7,2)
print(arr)
print("\n")
up_arr = arr.ravel()
up_arr[0] = 100
print(up_arr)
print("\n")
print(arr)
```

```python
#example ravel 3d
import numpy as np
arr = np.arange(14).reshape(7,2,1) # 2D blocks , rows , columns
print(arr)
print("\n")
up_arr = arr.ravel()
up_arr[0] = 100
print(up_arr)
print("\n")
print(arr)
```

```python
# transpose -> rows convert into cols || cols convert into rows
# transpose not possible in 1D bcz no rows & columns present in it
# example transpose 2d
import numpy as np
arr = np.arange(12).reshape(3,4)
print(arr)
print("\n")
up_arr = arr.T
print(up_arr)
```

```python
# example transpose 3d
# [block,row,col] --> [col,row,block]
import numpy as np
arr = np.arange(12).reshape(3,2,2)
print(arr)
print("\n")
up_arr = arr.T
print(up_arr)
```

**Numpy Array operations**

1) slicing for 1d,2d,3d

```python
#slicing for 1d
import numpy as np
arr = np.arange(11)
print(arr)
up_arr = arr[:3] # 0 -> 2 index
print(up_arr)

# [start : stop]
# Start = (included)
# Stop  = (excluded)
```

```python
#slicing for 2d
import numpy as np
arr1 = np.arange(16).reshape(8,2)
print(arr1)
up_arr1 = arr1[:,:-1]
print(up_arr1)

#[:, :-1]
# all rows , all colmns except the last one
# [rows,colmns]
```

```python
import numpy as np
arr1 = np.arange(16).reshape(8,2)
print(arr1)
up_arr1 = arr1[:,-1]
print(up_arr1)

#A single index removes that dimension.
# all rows and last column included
# makes it 1D
```

```python
#slicing for 3d
# create array --> arange(24) -> print 0,23

import numpy as np
arr2 = np.arange(24).reshape(3,4,2)
print(arr2)
print("\n")
# manually access --> this is not slicing
print(arr2[0,0,0],arr2[2,3,1])
# print(arr2[0,0,0],arr2[-1,-1,-1])
print("\n")

# slicing -> 0,23
print(arr2[::2, ::3, ::])
# this gives more elements along with it
# ::2 --> 0 to 2 included
# ::3 --> 0 to 3 included
# :: --> include all rows
```

**Looping with numpy**

1) while loop
2) for loop

```python
# while loop for 1d
import numpy as np
arr = np.arange(4)
i = 0
while i < len(arr):
  print(i, end=" ")
  i += 1

# ends =" "
# After printing, put a space instead of going to the next line.

# in for loop we take only arr in condn bcz it takes one value at a time
# whereas in case of while loop it takes complete array once due to which it gets confused
# so we take len(arr) to take only one value of it at once
```

```python
# while loop for 2d
import numpy as np
arr1 = np.arange(1,7).reshape(2,3)
print(arr1)
print("\n")
print(len(arr1)) # --> 2
print("\n")
i = 0
while i < len(arr1):
  #  print(len(arr1[i])) --> 3
  j = 0
  while j < len(arr1[i]):
    print(arr1[i][j], end=" ")
    j += 1
  print()
  i += 1

# len(arr) --> Number of rows = 2
# len(arr[i]) --> Number of columns in row i = 3
```

```python
# while loop for 3d
import numpy as np
arr = np.arange(8).reshape(2,2,2)
print(arr)
print("\n")
i = 0
while i < len(arr):
  j = 0
  while j < len(arr[i]):
    k = 0
    while k < len(arr[i][j]):
      print(arr[i][j][k], end = " ")
      k += 1
    print()
    j += 1
  print()
  i += 1
```

```python
# for loop 1d array

import numpy as np
arr = np.arange(4)
for i in arr:
  #Take each element from arr one by one and store it in i.
  # loop from 0 to 3 --> (arr -1)
    print(i, end =" ")
```

```python
# for loop 2d array

import numpy as np
arr = np.arange(4).reshape(2,2)
print(arr)
print("\n")
print(len(arr)) # 2
print("\n")
for row in arr:
    for value in row:
        print(value, end=" ")
    print() # prints 1 empty line space
```

```python
# for loop 3d array

import numpy as np

arr = np.arange(8).reshape(2,2,2)
print(arr)
print("\n")
for matrix in arr:
    # print(matrix) --> [[0 1]
                      #  [2 3]]
    for row in matrix:
        # print(row) --> [0 1]
        for value in row:
            # print(value) --> 0
            print(value, end=" ")
        print()
    print()
```
