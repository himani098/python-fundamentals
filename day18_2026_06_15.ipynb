**Handling missing values**

1) df.isnull() -> Check Missing Values in DataFrame
2) df.isnull().sum() -> Count Missing Values in Each Column
3) df.isnull().any() -> Check if Any Missing Value Exists
4) df.isnull().sum().sum() -> Count Total Missing Values
5) df.dropna() -> Remove Rows with Missing Values
6) df.dropna(axis=1) -> Remove Columns with Missing Values
7) df.fillna(0) | df["Salary"].fillna(0)-> Fill Missing Values with a Fixed Value
8) df["Salary"].fillna( df["Salary"].mean() ) -> Fill Missing Values with Mean
9) df["Salary"].fillna( df["Salary"].median() ) -> Fill Missing Values with Median
10) df["Department"].fillna( df["Department"].mode()[0] ) -> Fill Missing Values with Mode
(Since mode() returns a Series, we use [0] to get the first mode value.)

to handle a missing data we can either fill that value or delete that column or row if not necessary --> handling missing data --> data cleaning

| `df.isnull()`             | `df.isnull().any()`                          |
| ------------------------- | -------------------------------------------- |
| Shows every missing value | Checks if a column has **any** missing value |
| Returns a DataFrame       | Returns a Series                             |


isnull() --> True --> missing value

False --> no missing value

| Code                      | Output                                              |
| ------------------------- | --------------------------------------------------- |
| `df.isnull()`             | True/False for every cell                           |
| `df.isnull().any()`       | Checks if each column has any missing values        |
| `df.isnull().sum()`       | Counts missing values in each column                |
| `df.isnull().sum().sum()` | Counts total missing values in the entire DataFrame |


Mean: The average of all values.

Sum of all values ÷ Number of values.

Median: The middle value after arranging the data in ascending order.

Mode: The most frequently occurring value in the dataset.

| `NaN`                                  | `null` (`None`)                          |
| -------------------------------------- | ---------------------------------------- |
| Stands for **Not a Number**            | Means **no value / null**                |
| Provided by NumPy (`np.nan`)           | Python's built-in `None`                 |
| Commonly used for missing numeric data | Can represent missing values of any type |
| Type: `float`                          | Type: `NoneType`                         |


NaN doesn't mean it's a number. It means it's a special floating-point value used to represent a missing or undefined numeric value.

Numeric column → None becomes NaN.

Object/String column → None is stored as None.

If you're working with a JSON file, None is exported as null when you save it using to_json().

```python
import pandas as pd
import numpy as np
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/students_sample_02.json"
df = pd.read_json(url)
print(df)
print("\n")
# df.loc[4,"marks"] = None
df.loc[4,"marks"] = np.nan
df.loc[3,"roll no"] = np.nan
print(df)
print("\n")
print(df.isnull())
print("\n")
print(df.isnull().any())
print("\n")
print(df.isnull().sum())
print("\n")
print(df.isnull().sum().sum())
print("\n")
print(df.dropna()) #removes row
print("\n")
print(df.dropna(axis = 1)) #removes column
print("\n")
print(df.fillna(0))
print("\n")
print(df["marks"].fillna(0))
print("\n")
print(df["marks"].fillna(df["marks"].mean()))
print("\n")
print(df["marks"].fillna(df["marks"].mode()[0]))
```


**Aggregation and group by**

1) df["Salary"].sum() -> Calculate Sum
2) df["Salary"].mean() -> Calculate Average (Mean)
3) df["Salary"].max() -> Find Maximum Value
4) df["Salary"].min() -> Find Minimum Value
5) df["Department"].count() -> Count Values
6) df["Department"].value_counts() -> Count Values
7) df.groupby("Department")["Salary"].mean() -> Group By and Calculate Mean -->(group/combines the rows with same departments first then find the mean of their salary for the each department)
8) df.groupby("Department")["Salary"].sum() -> Group By and Calculate Sum
9) df.groupby( ["Department", "Gender"] )["Salary"].mean() -> Group By Multiple Columns --> (group first by department & then by gender next and then find the mean of their salary)
10) df.groupby("Department")["Salary"].agg( ["sum", "mean", "max", "min", "count"] ) -> Multiple Aggregations --> (Calculates multiple statistics for Salary.)
11) df.groupby("Department").agg({'col1':'sum','col2':'count'}) -> multiple aggregations

| Method                            | What it does                                              |
| --------------------------------- | --------------------------------------------------------- |
| `df["Department"].count()`        | Counts the **total number of non-null (non-NaN) values**. |
| `df["Department"].value_counts()` | Counts **how many times each unique value appears**.      |


```python
# aggregation
import pandas as pd
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/students_sample_02.json"
df = pd.read_json(url)
print(df)
print("\n")
# manually
print(df['marks'].sum())
print("\n")
print(df['marks'].mean())
print("\n")
print(df['marks'].min())
print("\n")
print(df['marks'].max())
print("\n")
print(df['marks'].count())
print("\n")
print(df['marks'].value_counts())
print("\n")
# usin aggregation
print(df.groupby("roll no")['marks'].agg(["sum","mean","min","max"]))
print("\n")
print(df['marks'].agg(["sum","mean","min","max"]))
print("\n")
print(df.agg({"roll no":"sum","marks":"count"}))
print("\n")
df.groupby("roll no")["marks"].mean()
```


**Concatenate and merge dataframe (joins)**

1) pd.concat([df1,df2],axis=0) -> row joins
2) pd.concat([df1,df2],axis=1) -> cols joins
3) pd.merge(df1,df2,how='inner',on='id') -> inner join basis of id

```python
# concatenate
import pandas as pd
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/students_sample_02.json"
df = pd.read_json(url)
df1 = pd.read_json(url)
print(pd.concat([df,df1],axis=0))
print("\n")
print(pd.concat([df,df1],axis=1))
print("\n")
pd.merge(df,df1,how='inner',on='marks')

# on='id' → Column used for matching.
# how='inner' → Type of merge (which rows to keep).

# Other how values
# inner → Keep only matching rows.
# left → Keep all rows from the left (df).
# right → Keep all rows from the right (df1).
# outer → Keep all rows from both DataFrames.
```
