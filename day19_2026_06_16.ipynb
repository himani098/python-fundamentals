What is matplotlib?

Matplotlib is a Python library used for creating static, animated, and interactive plots.

Data visualization: It's the graphical representation of data to identify patterns, trends, and insights.

pip install matplotlib
     
first we clean data with the help of numpy & pandas & then represent it in graph form

```python
#create first chart(line chart)
import matplotlib.pyplot as plt #visulaization
#Imports the pyplot module from the Matplotlib library and gives it the alias plt.
x = [2010, 2015, 2020, 2025] #x coordinate
y = [100, 200, 250, 300] # y coordinate
#same no of values should be in x & y both
plt.plot(x,y) #line graph
#Creates a line graph using x values for the x-axis and y values for the y-axis.
plt.show() #show graph
# Displays the graph in a separate window or below the code
```

Matplotlib → A Python library used for creating graphs, charts, and data visualizations.

Pyplot → A module inside the Matplotlib library that provides functions to create and display plots easily.

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.plot(x,y)
plt.xlabel("years") # x label
plt.ylabel("sales") # y label
plt.title("sales report") # graph title
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.figure(figsize = (6,3)) # graph size #(width, height)
plt.plot(x,y, color = "green", marker = 'o', linestyle = "dashed", linewidth = 4, markersize = 14,)
plt.show()
```

```python
#multiple line chart
import matplotlib.pyplot as plt
x = [2010, 2015, 2020, 2025]
y1 = [100, 200, 260, 290]
y2 = [150, 185, 195, 300]
plt.plot(x,y1,label = "jeans")
plt.plot(x,y2,label = "shirt")
plt.legend() #for side box of lines #info of label
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
a = [120, 150, 170, 160, 190, 210]
b = [100, 130, 140, 180, 170, 200]
c = [90, 110, 150, 170, 200, 230]
plt.figure(figsize = (6,4))
plt.plot(x,a,label = "product_a",color = "g", linestyle = "-", marker = "o")
plt.plot(x,b,label = "product_b",color = "k", linestyle = "--", marker = "h")
plt.plot(x,c,label = "product_c",color = "r", linestyle = "-.", marker = "*")
plt.legend()
plt.xlabel("Months")
plt.ylabel("sales")
plt.title("Monthly Sales")
plt.show()
```

```python
#bar chart
import matplotlib.pyplot as plt
x = [2010,2015,2020,2025]
y = [100,200,250,300]
plt.figure(figsize = (6,2))
plt.bar(x,y)
plt.show()
```

```python
# multiple bar chart
import matplotlib.pyplot as plt
import numpy as np
x = np.array([1,2,3,4])
y1 = [10, 20, 20, 40]
y2 = [20, 30, 25, 30]
w = 0.40
plt.figure(figsize = (6,4))
plt.bar(x - w/2,y1,label = "Girls", width = w)
plt.bar(x + w/2,y2,label = "Boys", width = w)
plt.legend()
plt.xlabel("Groups")
plt.ylabel("Number of Students")
plt.title("Number of Students in each group")
plt.show()
```
