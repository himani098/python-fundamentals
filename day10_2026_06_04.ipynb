**creating array from list**

1d

2d

```python
# list 1d
l = [1,2,3]
print(l)

import numpy as np
# array 1d
arr = np.array(l)
# arr -> variable name
# array() is a NumPy function used to create a NumPy array.
print(arr)
```

```python
# 2d example

import numpy as np
# 2d list
l = [[1,2,3],[4,5,6]]
print(l)
print(l[0])
print(l[0][0])
# l([row][column])

# question list value 4 -> 100
l[1][0] = 100
print(l[1])
print("\n")

# 2d array
arr = np.array([[1,2,3],[4,5,6]])
# only 1 argument can be passed so putted all lists in one bracket
print(arr)
print(arr[0])
print(arr[0][0])

# question array value 4 -> 100
arr[1][0] = 100
print(arr[1])
```

use of zeros and ones - if we have any missing value in a data then we can replace it by a zero or one or we can remove/delete it or we can take out it's mean and save it

```python
# zeros array 1d
#Creates an array filled with 0s.

import numpy as np
arr =  np.zeros(5)
print(arr)
print("\n")

# zeros array 2d
arr1 = np.zeros((3,4))
print(arr1)
```

```python
# ones array 1d
#Creates an array filled with 1s.

import numpy as np
arr = np.ones(5)
print(arr)
print("\n")

# ones array 2d
arr1 = np.ones((3,4))
print(arr1)
```

```python
#question
# create a 2D array with zeros and increment each element by 10

import numpy as np
arr = np.zeros((3,4))
print(arr)
print("\n")
print(arr+10)
```

```python
#question
# create a 2D array with ones and multiply each element by 5

import numpy as np
arr = np.ones((3,4))
print(arr)
print("\n")
print(arr*5)
```

```python
# full for 1d
#Used to create an array filled with the same value.
#np.full(shape, value)

import numpy as np
arr = np.full((3),5)
# 3 elements in 1D with value 5
print(arr)
print("\n")

# full for 2d
import numpy as np
arr1 = np.full((2,3),5)
# 2 rows & 3 columns in 2D with value 5
print(arr1)
```

```python
#question
# create a 2D array with zeros and value of each element as 5
# create a 2D array with full and value of each element as 1

import numpy as np

arr = np.zeros((3,4)) + 5
print(arr)
print("\n")

arr1 = np.full((3,4),1)
print(arr1)
```

```python
#random for 1d
#Used to generate random decimal numbers between 0 and 1

import numpy as np
arr = np.random.random(5)
print(arr)
print("\n")

# random for 2d
arr1 = np.random.random((2,3))
print(arr1)
```

```python
#arange for 1d
#Creates an array of numbers in a range.

import numpy as np
arr = np.arange(5) # from 0 to 4
print(arr)
print("\n")

#arange for 1d || to convert into 2d -> reshape
arr1 = np.arange(0,10,2) # start,stop,step
print(arr1)
```


**Vector,matrix,tensor**

1) vector (1d)
2) matrix (2d)
3) tensor (3d)

```python
import numpy as np

# vector 1d list
l = [1,2,3]
print(l)
print("\n")

# vector 1d array
arr = np.array(l)
print(arr)
print("\n")

# matrix 2d list
l = [[1,2,3],[4,5,6]]
print(l)
print("\n")

# matrix 2d array
arr = np.array(l)
print(arr)
print("\n")

# tensor 3d list
l = [[[1,2],[3,4]],[[5,6],[7,8]]]
print(l)
print("\n")

# tensor 3d array
arr = np.array(l)
print(arr)
```


**Array properties**

1) shape
2) dimension | direction
3) size
4) dtype

```python
# array
import numpy as np
arr = np.arange(12) # 1d
print(arr)
# print("shape: ",arr.shape)
print("shape:",np.shape(arr)) # in form of single row or rows & columns
# arr.shape → property/attribute
# np.shape(arr) → NumPy function

print("dimension: ",np.ndim(arr))
print("size: ",np.size(arr)) # total elements
print("datatype: ",arr.dtype)
# int64 -> All elements are integers
```

```python
# array
import numpy as np
arr = np.arange(12).reshape(3,4) # 3 X 4 => 12  # 2d
print(arr)
print("shape:",np.shape(arr))
print("dimension: ",np.ndim(arr))
print("size: ",np.size(arr))
print("datatype: ",arr.dtype)
```
