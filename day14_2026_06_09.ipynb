```python
#example 2
import pandas as pd
d = {"name":"kunal","age":"21","roll-no":101}
df = pd.Series(d)
# d is passed as the first positional argument (which is the data parameter)
print(df)
# string datatype --> object
```

```python
#example 2
import pandas as pd
d = {"name":"kunal","age":"21","roll-no":101}
df = pd.Series(data=d,index=["name","age","roll-no"])
# Here, you're explicitly naming the data parameter and also providing custom row labels
# Instead of 0, 1, 2, the rows are labeled "name", "age", and "roll-no"
# df = pd.Series(data=d,index=["0","1","2"])
print(df)
```

```python
import pandas as pd
d = {"name":"kunal","age":"21","roll-no":101}
df = pd.Series(data=d)
print(df)
```

```python
import pandas as pd
d = {"name":"kunal","age":"21","roll-no":101}
df = pd.Series(data=d,index=["0","1","2"])
# Since those keys don't exist in the dictionary, Pandas fills the values with NaN.
print(df)
```
