```python
#Bitwise operator
# >> , << , & , | , ^

# &
a = 5 # 101
b = 2 # 010
print(a & b)

# |
a = 5
b = 2
print(a | b)

# ^
a = 5
b = 2
print(a ^ b)

# >> -> right shift
a = 5
b = 2
print(a >> b) #101 --> 001

# << -> left shift
a = 5
b = 2
print(a << b) #101 --> 10100
```

**alternative for >> and <<**

a >> n   --> a // (2^n)

a << n   --> a * (2^n)

```python
print(22 // 2)
print(22 >> 1)
print(22 // 4)
print(22 >> 2)

print(22 * 2)
print(22 << 1)
print(22 * 4)
print(22 << 2)
```

```python
# Conditional Statements
# if
# elif
# else

# if
num1 = 10
num2 = 20
if num1 < num2 :
  print(num2)

# if else
num1 = int(input("enter the number: "))
if num1 :
  print("number:-",num1)
else:
  print("zero",num1)

# if elif else
age = int(input("Enter the number: "))
if age == 18:
  print("your age is: ",age)
elif age > 18 :
  print("you are above then 18 <",age)
else:
  print("you are less than 18 >",age)
```

```python
#loops
#for
#while
#break , continue, pass

#for
for i in range(5):
  #runs from 0 to 4
  # 0 is the by default starting value of it
  print("i = ",i)

#for
for i in range(1,11):
  # runs from 1 to 10
  print("number = ",i)


# list
# indexing starts from 0
# position and length starts from 1

l = [11,12,13,14]
print(len(l))
# len() -> used to find the length
for i in range(4):
  print(l[i])

# or
for i in range(len(l)):
  print(l[i])

# or
for i in l:
  print(i)
```

```python
#while
num1 = int(input("Enter the number: "))
i = 0
while i < num1 :
  print(i)
  i += 1

# list
l = [10,11,12,13,14,15]
i = 0
while i < len(l):
  print(l[i])
  i += 1

# example for | while
d = {"name":"helo","age":20,"phone":"00000000"}
# phone no writen in string bcz no calculation done on it
l = d.keys()
print(l)
for i in range(len(d.keys())):
  print(i)


# list example
l = [10,[11,12,13,14,15]]
print(l[1])

for i in l[1]:
  print(i)

print("second approch")

for i in range(2,len(l[1])):
   print(l[1][i])
```

```python
# break
for i in range(5):
  if i == 2:
    break
  print(i)


# continue
i = 0
while True:
  print(i)
  if i == 5:
    break #stops the loop there and don't run further
  else:
    i += 1
    continue #skips the current iteration and moves to the next one.

# pass
#when no statement given in the if block then it gives error
#so when we don't want to write any statement in it
#and we don't want error also then we write pass in it to avoid any error
i = 0
while True:
  print(i)
  if i == 4:
    pass
  if i == 5:
    break
  else:
    i += 1
    continue
```
