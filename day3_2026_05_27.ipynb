**Bitwise operator ->**
# >> , << , & , | , ^

```python
# &
a = 5 # 101
b = 2 # 010
print(a & b)
```

```python
# |
a = 5
b = 2
print(a | b)
```

```python
# ^
a = 5
b = 2
print(a ^ b)
```

```python
# >> -> right shift
a = 5
b = 2
print(a >> b) #101 --> 001
```

```python
# << -> left shift
a = 5
b = 2
print(a << b) #101 --> 10100
```

**alternative for >> and <<**

a >> n   --> a // (2^n)

a << n   --> a * (2^n)

```python
print(22 // 2)
print(22 >> 1)
print(22 // 4)
print(22 >> 2)
```

```python
print(22 * 2)
print(22 << 1)
print(22 * 4)
print(22 << 2)
```

```python
# break
for i in range(5):
  if i == 2:
    break
  print(i)
```

```python
# continue
i = 0
while True:
  print(i)
  if i == 5:
    break #stops the loop there and don't run further
  else:
    i += 1
    continue #skips the current iteration and moves to the next one.
    # also skips anything written after continue
```

```python
# pass
#when no statement given in the if block then it gives error
#so when we don't want to write any statement in it
#and we don't want error also then we write pass in it to avoid any error
i = 0
while True:
  print(i)
  if i == 4:
    pass
  if i == 5:
    break
  else:
    i += 1
    continue
```

```python
#HW
# question 1

# ATM PIN Validation

correct_pin = 23456
attempts = 3
for i in range(attempts):
  pin = int(input("Enter the pin: "))
  if correct_pin == pin :
    print("Login Successful")
    break;
  else :
    attempts -= 1
    print("Incorrect Pin")
    print("Attempts left:", attempts)
    if attempts == 0:
      print("Account Blocked")
    else:
      print("Try again...")
```

```python
# question 2

# Passwod Strength Checker

password = input("Enter the password: ")
uppercase = False
lowercase = False
digit = False
special_character = False

for ch in password:
    if ch.isupper():
        uppercase = True
    elif ch.islower():
        lowercase = True
    elif ch.isdigit():
        digit = True
    else:
        special_character = True

if len(password) < 8:
    print("Weak Password")

elif uppercase and lowercase and digit and special_character:
    print("Strong Password")

else:
    print("Medium Password")

#These are built-in string methods in Python.
#They check the type of characters in a string and return True or False.

# .isupper()
#Checks if a character/string is uppercase

# .islower()
#Checks if a character/string is lowercase.

# .isdigit()
#Checks if a character/string is a digit.

#The dot (.) is used to access something that belongs to an object.
#name = "HELLO"
#print(name.lower())
#name      → Object
#lower()   → Method belonging to that object
```

```python
#question 3

# Multiplication Table of a Number

num = int(input("Enter a number: "))
if num > 0:
    print("Multiplication Table:")
    for i in range(1, 11):
        print(num, "x", i, "=", num * i)
else:
    print("Please enter a positive number")
```

```python
#question 4

# Count Even and Odd Numbers from 1 to N

num = int(input("Enter the number:"))
even = 0
odd = 0
for i in range(1,num+1):
  if i % 2 == 0:
    even += 1
  else:
    odd += 1
print("Even Count =",even)
print("Odd Count =",odd)
```

```python
#question 5

#Number Guessing Game

secrete_number = 555
while True:
  guess = int(input("Enter the number:"))
  if guess == secrete_number:
    print("Correct Guess")
    print("You WIN !!")
    break
  elif guess < secrete_number:
    print("Guess Higher")
  else:
    print("Guess Lower")
```
