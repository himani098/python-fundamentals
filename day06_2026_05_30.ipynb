**Methods** - these are the functions which are object oriented and cann't be applied to all but to that particular object itself.
The dot is used to access methods and attributes of an object.

ex- functions of list like list.append()

**Functions** - these are the functions that can be applied to any object.

ex- len()

**Importing & Using Modules-**

way to use package / Library / modules :-

1) **import math**

imports the math library for use of it's functions.

ex - math.Pi()

Pi is the function of the math library which we are using here

2) **from math import Pi**

we are only importing the Pi function from the math library

ex - Pi()

3) **from math import** *

-> * means to import all the functions from the math library

ex - Pi()

4) **import math as raj**

import the math library but not as math but with the name raj

ex - raj.Pi()


```python
#import module
import math

print(math.factorial(5))
print(math.pow(5,2))
print(math.pi)
print(math.floor(10.7234))
print(math.ceil(10.0123))
#round a number UP to the nearest integer.
```

```python
# exmaple 1
import math as himani

print(himani.factorial(5))
print(himani.pow(5,2))
print(himani.pi)
print(himani.floor(10.7234))
print(himani.ceil(10.0123))
```

```python
# example 2
from math import factorial,pi

print(factorial(5))
print(pi)
```

```python
# example 3
from math import *

print(pow(5,3))
print(pi)
print(floor(10.3211))
```
