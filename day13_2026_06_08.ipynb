2) list -> indexing name (pd dataframe)

```python
import pandas as pd
df = pd.Series([10,20,30])
#A Series is a 1-dimensional labeled(indexed) array in Pandas.
#Pandas automatically adds indexes
print(df)
print(df[0])

# 0         10
# index    value
# dtype = data type || dtype: int64 (all values are integer with 64 bit space)
```

```python
# dict
# like for using the array methods we need to convert a list into array by using numpy
# like that only to use pandas methods we need to convert the dictionary into a data frame first
import pandas as pd
dict = {"name":["dheeraj","kunal","praveen","anjali","abhishek singh"],
        "age":[20,21,22,23,24],
        "salary":[{"In-hand":"15000","ctc":"1.2lpa"},{"In-hand":"20000","ctc":"2.2lpa"},
                  {"In-hand":"25000","ctc":"3.2lpa"},{"In-hand":"30000","ctc":"3.2lpa"},
                  {"In-hand":"40000","ctc":"4.2lpa"}
                  ]
      }
df = pd.DataFrame(dict)
print(df)

#DataFrame is a 2-dimensional table (rows and columns) in Pandas, similar to an Excel sheet.
```

```python
import pandas as pd
import os

url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/student_data.csv"
df = pd.read_csv(url)

df.to_csv("student_data.csv")      # Create CSV

df.to_json("student_data.json")    # Create JSON

df.to_excel("student_data.xlsx")   # Create Excel

print(os.listdir())
# if a file already exists & then we export it then it would rewrite that file
#Pandas needs another library called openpyxl to create Excel files.
#openpyxl is a Python library used to read, write, and modify Excel (.xlsx) files.
```
