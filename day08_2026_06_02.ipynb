```python
class Address:
  def __init__(self,city,state):
    self.city = city
    self.state = state

  def location(self):
    return f"my location is {self.city} in {self.state}"

#inheritance

class User(Address):
  def __init__(self,name,age,city,state):
    super().__init__(city,state)
    #used in inheritance to call the constructor of the parent class.

    self.name = name
    self.age = age

  def userName(self):
    print(f"my name is {self.name}")

  def userDetails(self):
    print(f"my name is {self.name} and my location is {self.city} in {self.state}")

u = User("himani",20,"kukas","rajasthan")
print(u.city)
u.userDetails()
u.location()
```

**Encapsulation**

```python
class Address:
  def __init__(self,city,state):
    self.city = city #attribute -> self.city
    self.state = state

  def location(self):
    print(f"my location is {self.city} in {self.state}")

a = Address("jaipur","rajasthan")
a.location()
print(a.city) #attribute -> a.city
print(a.state)
```

```python
class Address:
  def __init__(self,city,state):
    self.__city = city
    self.state = state

  def location(self):
    print(f"my location is {self.__city} in {self.state}")

a = Address("jaipur","rajasthan")
a.location()
# print(a.__city)
#AttributeError: 'Address' object has no attribute '__city'
#bcz it becomes private by putting double underscore(__)

# when we put only single underscore (_) -> '_city'
#then it doesn't give any error bcz it considers it as a variable only
print(a.state)
```

```python
class Address:
  def __init__(self,city,state):
    self.__city = city # private attribute
    self.state = state

  def location(self):
    print(f"my location is {self.__city} in {self.state}")

  def getcity(self):
    return self.__city

a = Address("jaipur","rajasthan")
a.location()
a.getcity()
# print(a.__city)
print(a.state)

# to access an private attribute we can only do it by creating another method and accessing through it
# private attributes can be accessed directly only inside the class and not outside the class
# outside the class can be accessed by using another method
```

**Polymorphism -**
Same method name with Different implementations

In polymorphism, the behavior usually differs because different objects belong to different classes.

```python
class Address:

  def __init__(self,city,state):
    self.city = city
    self.state = state

  def location(self):
    print(f"my address location is {self.city} in {self.state}")

class HomeTown:
  def __init__(self,city,state):
    self.city = city
    self.state = state

  def location(self):
    print(f"my home town location is {self.city} in {self.state}")

a = Address("vadodara","gujrat")
a.location()
b = HomeTown("jaipur","rajasthan")
b.location()
```

**overloading and overriding**

overloading - Same method name, different parameters.

overriding - Child class provides its own version of a parent class method.

```python
# overloading ->  not possible

def address(city,state): # paramter -> 2
  print(f"my address is {city} in {state}")

def address(city,state,country): #parameter -> 3
  print(f"my address is {city} at {state} in {country}")

# address("chappra","bihar")
# overloading possible nhi hai
# bcz in python it would consider that method which was created at the last

address("jaipur","rajasthan","india")
```

```python
# overriding

class Address:

  def __init__(self,city,state):
    self.city = city
    self.state = state

  def location(self):
    print("address locations")

class User(Address):

  def __init__(self,name,age,city,state):
    super().__init__(city,state)
    self.name = name
    self.age = age

  def location(self): # overriding
    print("user locations")

u = User("arya mains",20,"kukas","rajasthan")
u.location()
```
