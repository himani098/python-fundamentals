**arithematic opr ->**
# + - * / // % **

```python
# +
a = 10
b = 20
print(a + b) # 10 + 20
```

```python
# -
c = 20
d = 10
print(c - d) # 20 - 10
```

```python
# * -> asterisk -> multiplication
e = 10
f = 5
print(e * f) # 10 * 5
```

```python
# /
num_1 = 20
num_2 = 2
print(num_1 /  num_2) # float -> by default float value
```

```python
# //
num_3 = 20
num_4 = 2
print(num_3 // num_4) # floor division -> integer
```

```python
# % -> modulus
num_5 = 10
num_6 = 3
print(num_5 % num_6)
```

```python
# ** -> Exponentiation Operator (Power Operator).
num_7 = 5
num_8 = 3
print(num_7 ** num_8)
# ** -> for power
# a**b -> a to the power b -> a^b
```

```python
# example (even or odd)
num9 = int(input("Enter the number: "))
num10 = 2
print(num9 % num10) # 0 -> it's even || 1 -> it's odd
```

**Relational opr ->**

True or False
# == , < , > , >= , <= , !=

#<   Less Than

#>   Greater Than

```python
# == -> Equal To Operator (Equality Operator).
num1 = 10
num2 = 11
print(num1 == num2)
```

```python
# < -> less than
num1 = 10
num2 = 11
print(num1 < num2)
```

```python
# > -> greater than
num1 = 20
num2 = 10
print(num1 > num2)
```

```python
# >= -> greater than OR equal to
num1 = 20
num2 = 10
print(num1 >= num2)
#20 is greater than 10 OR equal to 10?
```

```python
# <= -> less than OR equal to
num1 = 20
num2 = 10
print(num1 <= num2)
```

```python
# != -> not equal
num1 = 20
num2 = 10
print(num1 != num2)
```

**Assignment opr ->**
# = , += , -= , *= , /= , //= , %=

```python
# = -> assing opr used to declare a variable

a = 10
b = a + 20
print(b)
```

```python
# += -> add and assign
a = 10
a += 20
print(a)
```

```python
# -=
a = 10
a -= 20
print(a)
```

```python
# *=
a = 10
a *= 20
print(a)
```

```python
# /=
a = 10
a /= 20
print(a)
```

```python
# %=
a = 10
a %= 20
print(a)
```

```python
# //=
a = 10
a //= 20
print(a)
```

**Logical opr ->**
# and , or , not

```python
# and
a = 10
b = 20
print(a and b)
```

```python
a = -10
b = 10
print(a and b)
print(a or b)

name1 = "hello"
name2 = ""
#gives this output for and operation which shows no result on screen but an empty space
print(name1 and name2)
print(name1 or name2)
```
