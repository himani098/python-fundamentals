**Sorting**

1) df.sort_values("column_name") -> Sort by Single Column
2) df.sort_values("Age", ascending=True) -> Sort in Ascending Order
3) df.sort_values("Age", ascending=False) -> Sort in Descending Order
4) df.sort_values(["col1", "col2"]) -> Sort Multiple Columns
5) df.sort_values( ["Age", "Salary"], ascending=[True, False] ) -> Multiple Columns with Descending Order

```python
import pandas as pd
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/student_data.json"
df = pd.read_json(url)
print(df)
print("\n")
# to make permanent do inplace = True
print(df.sort_values("name")) #by default ascending
# Python sorts strings using ASCII/Unicode values.
# first comes the uppercase letters and then the lowercase letters
print("\n")
print(df.sort_values("english", ascending = True)) #for ascending order
print("\n")
print(df.sort_values("english", ascending = False)) #for descending order
print("\n")
print(df.sort_values(by=["english"]))
print("\n")
print(df.sort_values(["english","maths"]))
print("\n")
print(df.sort_values(["english","maths"],ascending = [True,False]))
#for same values of english in ascending order we arrange the maths values in descending order
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/student_data.json"
df = pd.read_json(url)
print(df)
print("\n")
print(len(df)) # total rows
print("\n")

# add column total = py+math+english
df['total']  = df['english'] + df['maths'] + df['english']
print(df)
print("\n")

# add row
df.loc[14] = ["rajendra",'Male',50,80,80,210]
print(df)
print("\n")

# update column and update row
df["maths"] = df["maths"] + 50
print(df)
print("\n")

df.loc[1] = ["Anjali", "Female",20,40,60,120]
print(df)
print("\n")

# delete column
df.drop("total",axis=1,inplace = True)
print(df)
print("\n")

# delete row
df.drop(14,axis=0,inplace = True)
print(df)
print("\n")
```

```python
# a DataFrame stores columns, and each column contains a list of values.
import pandas as pd
df = pd.DataFrame({
    "Name": ["Vishal", "Virat", "Vineet"],
    "Age": [21, 22, 23]
})
print(df)
print("\n")

new_row = pd.DataFrame({
    "Name": ["Riya"],
    "Age": [22]
})
print(new_row)
print("\n")

print(pd.concat([df, new_row],ignore_index=False))
print("\n")
print(pd.concat([df, new_row],ignore_index=True))
# "Don't keep the old index numbers. Create new ones."
```


**Working with Date Value**

1) df['doj'] -> Add New Date Column
2) df['doj'].dtype -> Check Data Type of Date Column
3) df['doj'] = pd.to_datetime(df['doj']) -> Convert Object to Datetime
4) df['doj'].dtype -> Check Data Type of Date Column
5) pd.Timestamp.now() -> Get Current Date & Time
6) df['doj'].dt.year -> Extract Year
7) df['doj'].dt.month -> Extract Month
8) df['doj'].dt.day -> Extract Day
9) df[(df['doj'] >= '2024-01-01') & (df['doj'] <= '2024-12-31')] -> Filter Between Two Dates
10) df['doj'].dt.day_name() --> for extract week day
11) df['doj'] + pd.to_timedelta(20,unit='D')
12) df['doj'] + pd.to_timedelta('20 days')
timedelta --> used for performing operations on dates and time

```python
import pandas as pd
url = "https://raw.githubusercontent.com/himani098/python-fundamentals/main/students_sample_02.json"
df = pd.read_json(url)
print(df)
print("\n")
df['doj'] = ['2025-01-10','2025-02-10','2025-03-10','2025-04-10','2025-05-10']
print(df)
print("\n")
print(df['doj'].dtype) #dtype('O') --> object type --> string type
print("\n")

# convert string to date
df['doj'] = pd.to_datetime(df['doj'])
print(df)
print("\n")
print(df['doj'].dtype)   #datetime64[ns]
print("\n")
# df['doj'].dtype   #dtype('<M8[ns]') --> Date and Time (datetime)

# date = YYYY-MM-DD
# date operation
print(df['doj'].dt.year)
print("\n")
print(df['doj'].dt.month)
print("\n")
print(df['doj'].dt.day)
print("\n")
print(df['doj'].dt.day_name())
print("\n")

# 20 days
print(df['doj'] + pd.to_timedelta(20,unit='D')) #using unit
print("\n")
print(df['doj'] + pd.to_timedelta('20 days')) #using strings
print("\n")
print(pd.Timestamp.now())
#2026-07-05 06:24:08.785030 --> hours:minutes:seconds:microseconds
print("\n")

print(df[(df['doj'] >= '2025-01-01') & (df['doj'] <= '2025-03-31')])
```

int → stores numbers
